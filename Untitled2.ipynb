```python
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from transformers import DistilBertTokenizer, DistilBertModel
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np
```

### Dataset
https://www.kaggle.com/datasets/pratyushpuri/ai-vs-human-content-detection-1000-record-in-2025

```python
df = pd.read_csv("ai_vs_human.csv")
df
```

## Missing Values Control

```python

text_data = df["text_content"].tolist()
numeric_cols = [
    "word_count","character_count","sentence_count","lexical_diversity",
    "avg_sentence_length","avg_word_length","punctuation_ratio",
    "flesch_reading_ease","gunning_fog_index","grammar_errors",
    "passive_voice_ratio","predictability_score","burstiness","sentiment_score"
]
numeric_data = df[numeric_cols].values
labels = df["label"].values

```

```python
print(df[numeric_cols].isnull().sum()) 
print(df[numeric_cols].describe())      
```

```python

df["text_content"] = df["text_content"].fillna("")

```

```python
import numpy as np
df["text_content"] = df["text_content"].fillna("")
df[numeric_cols] = df[numeric_cols].replace([np.inf, -np.inf], np.nan)
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())


numeric_data = df[numeric_cols].values  

scaler = StandardScaler()
numeric_data = scaler.fit_transform(numeric_data)



```

```python
print("NaN var mı?:", np.isnan(numeric_data).any())
print("Inf var mı?:", np.isinf(numeric_data).any())
```

```python
X_train_text, X_test_text, X_train_num, X_test_num, y_train, y_test = train_test_split(
    text_data, numeric_data, labels, test_size=0.2
)
```

```python
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
bert_model = DistilBertModel.from_pretrained("distilbert-base-uncased")



```

```python
class MixedDataset(Dataset):
    def __init__(self, texts, nums, labels, tokenizer, max_len=64):
        self.texts = texts
        self.nums = torch.tensor(nums, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.float32)
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            truncation=True,
            padding='max_length',
            max_length=self.max_len,
            return_tensors='pt'
        )
        return {
            'input_ids': encoding['input_ids'].squeeze(0),
            'attention_mask': encoding['attention_mask'].squeeze(0),
            'numeric': self.nums[idx],
            'label': self.labels[idx]
        }
```

```python
train_dataset = MixedDataset(X_train_text, X_train_num, y_train, tokenizer)
test_dataset = MixedDataset(X_test_text, X_test_num, y_test, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=8)
```

```python
class BertWithNumeric(nn.Module):
    def __init__(self, bert_model, numeric_size):
        super(BertWithNumeric, self).__init__()
        self.bert = bert_model 
        self.numeric_dense = nn.Linear(numeric_size, 32)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.3)
        self.classifier = nn.Linear(self.bert.config.hidden_size + 32, 1)

    def forward(self, input_ids, attention_mask, numeric):
        bert_out = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_token = bert_out.last_hidden_state[:, 0, :]
        num_out = self.relu(self.numeric_dense(numeric))
        combined = torch.cat((cls_token, num_out), dim=1)
        combined = self.dropout(combined)
        return self.classifier(combined)
```

### Eğitim Döngüsü

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BertWithNumeric(bert_model, numeric_size=len(numeric_cols)).to(device)

criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-5)

train_losses, val_losses, val_accuracies = [], [], []

for epoch in range(3):
    # ---- Eğitim ----
    model.train()
    total_loss = 0
    for batch in train_loader:
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        numeric = batch['numeric'].to(device)
        labels = batch['label'].float().unsqueeze(1).to(device)

        outputs = model(input_ids, attention_mask, numeric)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_train_loss = total_loss / len(train_loader)
    train_losses.append(avg_train_loss)


    model.eval()
    total, correct, val_loss = 0, 0, 0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            numeric = batch['numeric'].to(device)
            labels = batch['label'].float().unsqueeze(1).to(device)

            outputs = model(input_ids, attention_mask, numeric)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

            preds = (torch.sigmoid(outputs) > 0.5).float()
            correct += (preds == labels).sum().item()
            total += labels.size(0)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    avg_val_loss = val_loss / len(test_loader)
    val_losses.append(avg_val_loss)
    val_accuracy = correct / total
    val_accuracies.append(val_accuracy)

    print(f"Epoch {epoch+1} | "
          f"Train Loss: {avg_train_loss:.4f} | "
          f"Val Loss: {avg_val_loss:.4f} | "
          f"Val Acc: {val_accuracy:.4f}")



```

```python

plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("Loss Grafiği")
plt.show()


plt.plot(val_accuracies, label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Accuracy Grafiği")
plt.show()
```

```python
model.eval()
correct, total = 0, 0
all_preds, all_labels = [], []
with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        numeric = batch['numeric'].to(device)
        labels = batch['label'].float().unsqueeze(1).to(device)

        outputs = model(input_ids, attention_mask, numeric)
        preds = (torch.sigmoid(outputs) > 0.5).float()
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

print(f"Accuracy: {correct/total:.2f}")


cm = confusion_matrix(all_labels, all_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```
